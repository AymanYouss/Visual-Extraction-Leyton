```python
import cv2
import pytesseract
import numpy as np
import json

pytesseract.pytesseract.tesseract_cmd = "C:/Program Files/Tesseract-OCR/tesseract.exe"  # your path may be different

# Set the path to your image
image_path = 'image.png'

# Load the image
image = cv2.imread(image_path)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert image to grayscale

# Extract text using pytesseract
extracted_text = pytesseract.image_to_string(gray_image, lang='eng')

# Find contours for detecting shapes and tables
contours, _ = cv2.findContours(cv2.Canny(gray_image, 100, 200), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Placeholder for shapes
shapes = {
    "rectangles": [],
    "circles": [],
    # Add more shapes as needed
}

# Analyze contours to find shapes (rudimentary shape detection)
for cnt in contours:
    approx = cv2.approxPolyDP(cnt, 0.01 * cv2.arcLength(cnt, True), True)
    x, y, w, h = cv2.boundingRect(approx)
    if len(approx) == 4:  # Rectangles (including squares)
        shapes['rectangles'].append({'position': (x, y, x+w, y+h)})
    # Add more shape detections as needed

# For simplicity, we're assuming 'tables' are just recognized as multiple adjacent rectangles.
# Real table detection would be more complex and might require additional libraries or custom algorithms.

# Compile all extracted information
extracted_info = {
    "text": extracted_text,
    "visual_information": {
        "tables": {
            "Number of tables": "Assumed based on rectangles - update this as needed",
            "Position of tables": "Update this as needed based on your logic"
        },
        "geometrical_forms": shapes,
        # Add other interesting information as needed
        "contour_counts": len(contours),  # Example additional info
    }
}

# Convert to JSON
json_output = json.dumps(extracted_info, indent=4)

# Print or save the JSON output
print(json_output)
```

```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
import json
import io

# Set the path to your PDF file
pdf_path = 'cv.pdf'

# Open the PDF
doc = fitz.open(pdf_path)

# Initialize a structure to hold your extracted information
extracted_info = {
    "text": [],
    "visual_information": []
}

# Process each page in the PDF
for page_number in range(len(doc)):
    page = doc.load_page(page_number)
    text = page.get_text("text")
    extracted_info["text"].append(text)

    # Render page to an image
    pix = page.get_pixmap()
    img = Image.open(io.BytesIO(pix.tobytes()))

    # Use OCR to extract text (in case PDF text extraction misses something, e.g., in scanned documents)
    ocr_text = pytesseract.image_to_string(img, lang='eng')
    
    # Here we're simply adding the OCR text to our structure, but in a real case, you might want to compare or combine it with the PDF extracted text
    extracted_info["text"].append(ocr_text)

    # Placeholder for visual information - this part is highly customizable depending on your needs
    visuals = {
        "page_number": page_number + 1,
        # Example visual information; actual detection logic would be needed for real use cases
        "image_count": len(page.get_images(full=True)),
        # Add more as needed
    }
    extracted_info["visual_information"].append(visuals)

# Convert to JSON
json_output = json.dumps(extracted_info, indent=4)

# Print or save the JSON output
print(json_output)
```
